```python
import json
import matplotlib.pyplot as plt
import math
```

```python
with open('../Resources/subway_delays_map_data.txt', "r") as read_file:
        subway_file = json.load(read_file)
        top_10_delays = subway_file[0:11]
        for subway in top_10_delays:
            name = subway[0]
            value = subway[1]
            location = subway[2]
            px=plt.text(location['lng'],location['lat'], value, fontsize=10)
            ax=plt.plot(location['lng'],location['lat'],marker='o',color='Red',markersize=int(math.sqrt(value))*0.9)
```
