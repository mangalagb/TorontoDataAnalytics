```python
import json
import matplotlib.pyplot as plt
import math
```

```python
with open('../Resources/subway_delays_map_data.txt', "r") as read_file:
        subway_file = json.load(read_file)
        for subway in subway_file:
            name = subway[0]
            value = subway[1]
            location = subway[2]
            ax=plt.plot(location['lng'],location['lat'],marker='o',color='Red',markersize=int(math.sqrt(value))*0.5)
```
